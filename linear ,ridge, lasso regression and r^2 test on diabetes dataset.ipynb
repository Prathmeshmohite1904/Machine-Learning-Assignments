```python
from sklearn.datasets import load_diabetes
```

```python
import pandas as pd
import numpy as np
```

```python
df=load_diabetes()
df
```

```python
dataset=pd.DataFrame(df['data'])
```

```python
dataset
```

```python
dataset.columns=df['feature_names']
```

```python
dataset
```

```python
dataset['price']=df['target']
dataset
```

```python
## converting dataset into dependent and independent features
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
dataset.isnull()
```

```python
ds=dataset.dropna()
ds
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
lin_reg=model.fit(x_train,y_train)

y_pred=lin_reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score1=r2_score(y_pred,y_test)
score1
```

```python
from sklearn.model_selection import cross_val_score
mse=cross_val_score(lin_reg,x_train,y_train)
mse
```

```python
np.mean(mse)
```

```python
## Ridge
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[1e-15,1e-9,1e-1,4,8,14,48,72,93,100]}
ridge_regressor=GridSearchCV(ridge,params,cv=10)
ridge_regressor.fit(x,y)
```

```python
y_pred=ridge_regressor.predict(x_test)
y_pred
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
## lasso

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
model=Lasso()
params={'alpha':[1e-12,1e-4,1e-2,34,25,56,78,90,100,110]}
lasso_regressor=GridSearchCV(model,params,cv=10)
```

```python
lasso_regressor.fit(x_train,y_train)
```

```python

y_p=lasso_regressor.predict(x_test)
```

```python
y_p
```

```python
from sklearn.metrics import r2_score
r2score1=r2_score(y_p,y_test)
```

```python
r2score1
```

```python
from sklearn.model_selection import cross_val_score
```

```python
mse=cross_val_score(lasso_regressor,x_train,y_train)
```
