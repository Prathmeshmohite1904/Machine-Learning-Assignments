## Linear regression , Ridge and Lasson


```python
# iris dataset
from sklearn.datasets import load_iris
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=load_iris()
```

```python
df
```

```python
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
dataset
```

```python
dataset['price']=df.target
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,:-1]    ###converting dataset into dependent and independent features
y=dataset.iloc[:,-1]
```

```python
x   #independent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
     x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()

mse=cross_val_score(lin_reg,x_train,y_train,cv=5)
print(mse)
```

```python
mse_mean=np.mean(mse)
mse_mean
```

```python
##Linear Rgression .... mse=minm squared error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
mse=cross_val_score(lin_reg,x,y,cv=5)
print(mse)
```

```python
mse_avg=np.mean(mse)
mse_avg
```

```python
y   #dependent
```

```python
lin_reg.fit(x,y)
lin_reg.predict(x)
```

```python
##ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV    ##for hyperparametric tunning
ridge=Ridge()
params={'alpha':[1e-8,1e-3,12,23,78,100]}
ridge_regressor=GridSearchCV(ridge,params,cv=5)
ridge_regressor.fit(x,y)

```

```python
print(ridge_regressor.best_score_)
```

```python
print(ridge_regressor.best_params_)
```

```python
##lasso regresssion
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-12,1e-2,12,14,19,88,100]}
lasso_regressor=GridSearchCV(lasso,params,cv=5)
lasso_regressor.fit(x_train,y_train,)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

## Calculating R2 value for each regression model

```python
y_pred=lasso_regressor.predict(x_test)
from sklearn.metrics import r2_score
r2_score1=r2_score(y_pred,y_test)
print(r2_score1)
```

```python
y_pred=ridge_regressor.predict(x_test)
from sklearn.metrics import r2_score
r2_score1=r2_score(y_pred,y_test)
print(r2_score1)
```

```python
y_pred=lin_reg.predict(x_test)

from sklearn.metrics import r2_score
r2_score1=r2_score(y_pred,y_test)
print(r2_score1)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
```

```python
df
```

```python
import pandas as pd
```

```python
dataset=pd.DataFrame(df['data'],columns=df['feature_names'])
dataset
```

```python
dataset.columns=df['feature_names']
dataset
```

```python
dataset['dep_feature']=df['target']
```

```python

dataset
```

```python
x=dataset.iloc[:,:-1]  ####independent features

y=dataset.iloc[:,-1]   ###dependent feature
```

```python
print(x)
```

```python
print(y)
```

```python
y.value_counts()     ## To check dataset is balanced or not
```

```python
## Train test split

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1=LogisticRegression(C=5,max_iter=100)
```

```python
model=GridSearchCV(model1,param_grid=params,cv=5)
```

```python
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
